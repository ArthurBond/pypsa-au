```python
def new_append(lst,dct,key):
    try:
        lst.append(dct[key])
    except KeyError:
        lst.append('null')

station_id,sup_code,name,location_id,approved,description,created_by,created_at=[],[],[],[],[],[],[],[]

unit_id,network,fueltech,status,sub_code,dispatch_type,active,capacity_registered=[],[],[],[],[],[],[],[]
network_region,unit_number,unit_capacity,emissions_factor_co2,unit_approved=[],[],[],[],[]

for sup_i, gen in enumerate(data['data']):
    station_id_val = gen['id']
    sup_code_val = gen['code']
    name_val = gen['name']
    location_id_val = gen['location_id']
    approved_val = gen['approved']
    try:
        description_val = gen['description']
    except KeyError:
        description_val = 'null'
    created_by_val = gen['created_by']
    created_at_val = gen['created_at'] 

    if not gen['facilities']:
        print(sup_i,sub_i)

        print(gen['approved'])

        print('None')
    else:
        for sub_i, facility in enumerate(gen['facilities']):
            # print(sup_i,sub_i)
        
            new_append(unit_id,facility,'id')
            new_append(network,facility,'network')
            new_append(fueltech,facility,'fueltech')
            new_append(status,facility,'status')
            new_append(sub_code,facility,'code')
            new_append(dispatch_type,facility,'dispatch_type')
            new_append(active,facility,'active')
            new_append(capacity_registered,facility,'capacity_registered')
            new_append(network_region,facility,'network_region')
            new_append(unit_number,facility,'unit_number')
            new_append(unit_capacity,facility,'unit_capacity')
            new_append(emissions_factor_co2,facility,'emissions_factor_co2')
            new_append(unit_approved,facility,'approved')

            station_id.append(station_id_val)
            sup_code.append(sup_code_val)
            name.append(name_val)
            location_id.append(location_id_val)
            approved.append(approved_val)
            description.append(description_val)
            created_by.append(created_by_val)
            created_at.append(created_at_val)


```
